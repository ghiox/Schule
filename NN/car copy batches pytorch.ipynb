```python
import pandas as pd

df = pd.read_csv('Car details v3.csv')

df
```

```python
# keep first and second string before space in name column
df['name'] = df['name'].str.split(' ').str[0] + ' ' + df['name'].str.split(' ').str[1]


# remove kmpl from mileage column
df['mileage'] = df['mileage'].str.split(' ').str[0]

# remove bhp from engine column
df['engine'] = df['engine'].str.split(' ').str[0]

# remove CC from engine column
df['engine'] = df['engine'].str.split(' ').str[0]

# remove bhp from max_power column
df['max_power'] = df['max_power'].str.split(' ').str[0]

# drop torque column
df = df.drop('torque', axis=1)
df
```

```python
# transform name, fuel, seller_type, transmission, owner columns to numeric values
df['name'] = df['name'].astype('category')
df['name'] = df['name'].cat.codes

df['fuel'] = df['fuel'].astype('category')
df['fuel'] = df['fuel'].cat.codes

df['seller_type'] = df['seller_type'].astype('category')

df['seller_type'] = df['seller_type'].cat.codes

df['transmission'] = df['transmission'].astype('category')
df['transmission'] = df['transmission'].cat.codes

df['owner'] = df['owner'].astype('category')
df['owner'] = df['owner'].cat.codes

df.info()
```

```python
# convert mileage, engine, max_power to numeric values
df['mileage'] = pd.to_numeric(df['mileage'])
df['engine'] = pd.to_numeric(df['engine'])
df['max_power'] = pd.to_numeric(df['max_power'])

df.dtypes

df.info()

# sum of null values in each column
df.isnull().sum()

# fill null values with mean
df['mileage'] = df['mileage'].fillna(df['mileage'].mean())
df['engine'] = df['engine'].fillna(df['engine'].mean())
df['max_power'] = df['max_power'].fillna(df['max_power'].mean())
df['seats'] = df['seats'].fillna(df['seats'].mean())

df.info()
```

```python
# correlation heat map
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
sns.heatmap(df.corr(), annot=True, cmap="RdYlGn")
plt.show()
```

```python
y = df['selling_price']

# use log on y to reduce skewness
import numpy as np
y = np.log(y)

X = df.drop('selling_price', axis=1)
```

```python
# standardize data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(X)
# transform to dataframe
x = pd.DataFrame(x_scaled, columns=X.columns)
x
```

```python
# train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

```python
x.shape
```

```python
import torch

x_train_tensor = torch.tensor(x_train.values, dtype=torch.float)
x_test_tensor = torch.tensor(x_test.values, dtype=torch.float)
```

```python
import numpy as np

def root_mean_squared_error(y_hat, y):
    return torch.sqrt(torch.mean((y_hat - y)**2))

def get_batch(x,y,batch_size):
    for i in range(0,len(x),batch_size):
        # yield: return a generator
        yield x[i:i+batch_size],y[i:i+batch_size]
```

```python
import torch

class MyModel(torch.nn.Module):
    def __init__(self):
        super(MyModel, self).__init__()
        self.fc1 = torch.nn.Linear(11, 15)
        self.fc2 = torch.nn.Linear(15, 1)
        self.batch_norm1 = torch.nn.BatchNorm1d(15)
        self.relu= torch.nn.ReLU()
        
    def forward(self, xb):
        xb = self.relu(self.fc1(xb))
        xb = self.batch_norm1(xb)
        xb = self.fc2(xb)
        return xb
```

```python
lr=0.001
wd=0.1
torch.manual_seed(1234)

batch_size=10
epochs=100
losses=[]
rmse_test=[]

model = MyModel()

y_train_tensor=torch.tensor(y_train.values, dtype=torch.float32)
x_test_tensor=torch.tensor(x_test.values, dtype=torch.float32)
y_test_tensor=torch.tensor(y_test.values, dtype=torch.float32)
for i in range(epochs):
    losses_in_epoch=[]
    # print the epoch number every 100 epochs
    if i%100==0:
        print(i)
    # trainiere jeweils mit einer kleinen Auswahl an Daten, der Batch
    # xb: ist die Auswahl der Daten für die Inputneuronen
    # yb: ist die Auswahl der Daten für die Outputneuronen
    for xb,yb in get_batch(x_train_tensor,y_train_tensor,batch_size):
        y_hat=model(xb)
        loss=root_mean_squared_error(y_hat, yb)+ wd * sum([torch.sum(param**2) for param in model.parameters()])
        losses_in_epoch.append(loss.item())
        loss.backward()
        with torch.no_grad():
            for param in model.parameters():
                param-=param.grad*lr
            model.zero_grad()
    losses.append(sum(losses_in_epoch)/len(losses_in_epoch))
    y_test_hat=model(x_test_tensor)
    y_test_rmse=root_mean_squared_error(y_test_hat, y_test_tensor)
    rmse_test.append(y_test_rmse.item())

plt.plot(range(epochs),losses, color="blue")
plt.plot(range(epochs),rmse_test, color="red")
plt.legend(["loss", "rmse_test"])
plt.show()
```
