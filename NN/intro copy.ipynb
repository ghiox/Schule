```python
# create a dataframe with the data
import pandas as pd

df = pd.DataFrame(data["data"], columns=data["feature_names"])

df.head()
```

```python
y = df["alcohol"]
x = df.drop("alcohol", axis=1)
```

```python
x
```

```python
# take x data and normalize it
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaler.fit(x)
x_scaled = scaler.transform(x)
# transform x back into a dataframe
x = pd.DataFrame(x_scaled, columns= x.columns)
x.head()
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=42, test_size=0.2)
```

# Netzarchitektur
Ein paar Gedanken zur Architektur

* shape von x: 178, 12 --> 178 rows und 12 columns
* Wieviele Input-Neuronen brauchen wir? --> 12
* Wieviele Output-Neuronen brauchen wir? --> 1

Hiddenlayer je nach Geschmack. z. B. ein Hiddenlayer mit 30 Neuronen

```python
# Erstmal bauen wir uns x für das neuronale Netz
# das neuronale Netz kann kein Dataframe verarbeiten, sondern nur einen Tensor
import torch

x_train_tensor = torch.tensor(x_train.values, dtype=torch.float32)
x_train_tensor.dtype
```

## w1

Welches shapr braucht w1?
w1 geht von 12 neuronen auf 30 neuronen, das heißt w1 braucht den shape (12, 30)

```python
w1 = torch.normal(std=1, mean=0, dtype=torch.float32, size=(12, 30), requires_grad=True)

# convert to float64
```

## w2
Welches shape braucht w2?
w2 geht von 30 Neuronen auf 1 Neuron --> Shape (30, 1)

```python
w2 = torch.normal(std=1, mean=0, size=(30, 1), requires_grad=True, dtype=torch.float32)
```

```python
import numpy as np
def relu(x):
    return torch.max(torch.tensor(0.0), x)
```

```python
y_hat = relu(relu(x_train_tensor@w1)@w2)
y_hat
```

```python
y_train_tensor = torch.tensor(y_train.values)

loss = (y_train_tensor- y_hat)**2
loss
```

```python
loss.mean().backward()
loss
```

```python
w1.grad
```

```python
lr = 0.001
losses = []

b1 = torch.tensor(1.0, requires_grad=True)
b2 = torch.tensor(1.0, requires_grad=True)	

def train(x_train_tensor, y_train_tensor, w1, w2, lr, b1, b2):
    y_hat = relu(relu(x_train_tensor@w1+b1)@w2+b2)
    loss = (y_train_tensor- y_hat)**2
    loss.mean().backward()
    with torch.no_grad():
       
        w1.sub_(lr*w1.grad)
        w2.sub_(lr*w2.grad)
        w1.grad.zero_()
        w2.grad.zero_()
    return loss.mean().item()

for i in range(1000):
    loss = train(x_train_tensor, y_train_tensor, w1, w2, lr, b1, b2)
    losses.append(loss)

import matplotlib.pyplot as plt
plt.plot(losses)
```
